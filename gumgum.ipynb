```python
import pandas as pd

headers = ['id', 'timestamp', 
           'type', 'visitorId',
          'pageUrl']

df = pd.read_csv('ad-events-2018060100.tar.gz', names=headers, engine='c')

df.head()
```

```python
# Clean up values

for col in df.columns:
    df[col] = df[col].str.partition(':"')[2].str.strip('"}')
```

```python
df.head()
```

```python
# Parse date time object
df['timestamp'] = pd.to_datetime(df.timestamp, errors='coerce', unit='ms')
```
